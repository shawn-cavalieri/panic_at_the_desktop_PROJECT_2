```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data


mydata = data.DataReader("^IXIC",
                       start='1997-1-1',
                       end='2002-12-31',
                       data_source='yahoo')['Adj Close']
normalize = (mydata / mydata.iloc[0] * 100)
normalize.plot(figsize = (15, 6))
plt.show()
```
